```python
class Celular:
    # Método constructor para inicializar atributos
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    # Método para encender el celular
    def encender(self):
        self.encendido = True
        print(f"El {self.marca} {self.modelo} está encendido.")

    # Método para mostrar información
    def mostrar_info(self):
        return f"Celular: {self.marca} {self.modelo}, Precio: ${self.precio}"

# Creación de instancias (objetos)
celular1 = Celular("Samsung", "Galaxy S21", 700)
celular2 = Celular("Apple", "iPhone 13", 900)
# llamdas a las funciones para cada objeto celular 1 y celular 2
print("celular 1:")
celular1.encender()
print(celular1.mostrar_info())

# Uso de métodos
print("\ncelular 2:")
celular2.encender()
print(celular2.mostrar_info())

print("programa terminado")
print("creador de software por Axel Guzman NL= 23")
```
